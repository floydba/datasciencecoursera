# This is an example of running DSX tools against local Netezza environment. It will load several packages from GitHub that are necessary for analysis and then connect to the Netezza environment and run the analysis.


### Information on packages can be found here: 

- DatabaseConnector - https://github.com/aaron0browne/DatabaseConnector/tree/netezza-support
- SqlRender - https://github.com/aaron0browne/SqlRender/tree/netezza-support
- Achilles - https://github.com/OHDSI/Achilles

```r
installed.packages()
```

```r
install_github("aaron0browne/SqlRender")
```

```r
install_github("aaron0browne/DatabaseConnector")
```

```r
install_github("ohdsi/Achilles")
```
